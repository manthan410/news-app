# Text Summarisation

Two types: Extractive and Abstractive

Extractive summarization involves selecting the most important sentences or phrases from the original text to create a summary. It can be faster and more straightforward than abstractive summarization, but it may not capture the nuances or broader context of the original text.

Abstractive summarization involves generating new sentences to capture the meaning of the original text. It has the potential to create more comprehensive and readable summaries, but it can also be more challenging to implement and may require more training data.

For a news app that requires a brief summary of the news article, extractive summarization may be more suitable as it provides a concise summary that covers the main points of the article.

```python
# Preprocessing the text data
nlp = spacy.load('en_core_web_sm')
stop_words = spacy.lang.en.stop_words.STOP_WORDS
#lemmatizer = nlp.vocab.morphology.lemmatizer
def preprocess(text):
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text)

    # Remove special characters
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)

    doc = nlp(text)
    tokens = [token.lemma_.lower() for token in doc if not token.is_stop and not token.is_punct and not token.is_space]
    return tokens
```

```python
# Creating dictionary and corpus for LDA
#texts = df_politics['content'].apply(preprocess)
#texts = df['content_en'][0]
texts =df['content_en'].apply(preprocess)
dictionary = gensim.corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
texts.head()
```

```python
len(texts)
```

```python
# Running LDA model with 10 topics
num_topics = 11
topic_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                            id2word=dictionary,
                                            num_topics=num_topics,
                                            random_state=42,
                                            passes=100,
                                            per_word_topics=True)
```

```python
# check topic number for each article
topic_model[corpus[15]]
```

```python
# Print top topics and their respective keywords
topics = topic_model.show_topics(num_topics=10, num_words=10, formatted=False)
for topic in topics:
    print('Topic {}: {}'.format(topic[0], ', '.join([word[0] for word in topic[1]])))
```

```python
# Visualize the LDA model
lda_display = gensimvis.prepare(topic_model, corpus, dictionary, sort_topics=False)
pyLDAvis.display(lda_display)
```

```python
assigned_topics=[]
for text in df['content_en'].apply(preprocess):
    dictionary = gensim.corpora.Dictionary([text])
    corpus = [dictionary.doc2bow(text)]
    #print(corpus)
    num_topics = 3
    lda_model = gensim.models.LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics, random_state=42, passes=10)

    # Get the assigned topics for the document
    assigned_topic = lda_model[corpus[0]]

    # Append the assigned topics to the list
    assigned_topics.append(assigned_topic)
```

```python
#assigned_topics=[]
# Get unique categories
categories = df['category'].unique()
```

```python
categories
```

```python
for category in categories:
    # Filter dataframe for the current category
    category_df = df[df['category'] == category]
    
    assigned_topics = []
    dominant_topics = []

    # Iterate over articles in the current category
    for text in category_df['content_en'].apply(preprocess):
        # Create dictionary and corpus for LDA
        dictionary = gensim.corpora.Dictionary([text])
        corpus = [dictionary.doc2bow(text)]
        
        num_topics = 3
        lda_model = gensim.models.LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics, random_state=42, passes=10)
        
        # Get the assigned topics for the document
        assigned_topic = lda_model[corpus[0]]
        
        # Append the assigned topics to the list
        assigned_topics.append(assigned_topic)
        
        # Determine the dominant topic for the article
        dominant_topic = max(assigned_topic, key=lambda x: x[1])[0]
        dominant_topics.append(dominant_topic)
    
    # Determine the dominant topic for each article in the current category
    #dominant_topics = [max(topics, key=lambda x: x[1])[0] for topics in assigned_topics]
    
    df.loc[df['category'] == category, 'dominant_topic'] = dominant_topics
    

    
    # Perform LDA visualization
    texts = category_df['content_en'].apply(preprocess).tolist()
    dictionary = gensim.corpora.Dictionary(texts)
    corpus = [dictionary.doc2bow(text) for text in texts]
    
    lda_model = gensim.models.LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics, random_state=42, passes=10)
    
    # Generate the LDA visualization
    vis_data = gensimvis.prepare(lda_model, corpus, dictionary, sort_topics=True)
    
    # Save the visualization as an HTML file
    vis_file_path = f'{category}_lda_visualization.html'
    pyLDAvis.save_html(vis_data, vis_file_path)

```
