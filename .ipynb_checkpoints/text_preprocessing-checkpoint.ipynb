```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv

kompas = requests.get('https://www.kompas.com/')
beautify = BeautifulSoup(kompas.content,'html5lib')

news = beautify.find_all('div', {'class','most__list clearfix'})
arti = []
for each in news:
  nu = each.find('div', {'class','most__count'}).text
  title = each.find('h4', {'class','most__title'}).text
  lnk = each.a.get('href')
  rcount = each.find('div', {'class','most__read'}).text
  r = requests.get(lnk)
  soup = BeautifulSoup(r.text,'html5lib')
  content = soup.find('div', class_ = "read__content").text.strip()
  #print(nu)
  #print(title)
  #print(lnk)
  #print(rcount)

  arti.append({
    'Top Number': nu,
    'Headline': title,
    'Link': lnk,
    'Most Read': rcount,
    'Content':content
  })

df = pd.DataFrame(arti)
df.head()
#df.to_csv('kompas.csv', index=False)
```

```python
import newspaper
from newspaper import Article
from newspaper import Source
import pandas as pd
import re

paper = newspaper.build("https://www.t-online.de/nachrichten/ausland/", memoize_articles = False) 

#regex = r"(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'\".,<>?«»“”‘’]))"

title=[]
text=[]
pubdate=[]
source=[]
image=[]

for each_article in paper.articles[0:10]:
    each_article.download()
    each_article.parse()
  
    title.append(each_article.title)
    text.append(each_article.text)
    pubdate.append(each_article.publish_date)
    source.append(each_article.url)
    image.append(each_article.top_image)
df= pd.DataFrame({'title':title,'link':source ,'content':text,'pubDate':pubdate, 'image_url':image})

  
#final_df.to_csv('articles.csv')
```

```python
import requests
from bs4 import BeautifulSoup

url='https://www.spiegel.de/ausland/'
response = requests.get(url)
soup = BeautifulSoup(response.content,'html5lib')
#print(soup.prettify())
#news = beautify.find_all('div', {'class','bg-white dark:bg-dm-shade-darkest shadow rounded'})
news = soup.find_all('article', class_='lg:py-24 md:py-24 sm:py-16')
#print(news)
#print(news)
link=[]
title=[]
img=[]
content=[]
date=[]
for each in news[0:10]:
    title.append(each.a.get('title'))
    url1=each.a.get('href')
    link.append(url1)
    response1= requests.get(url1)
    soup1 = BeautifulSoup(response1.text,'html5lib')
    #print(soup1)
    
    #soup = BeautifulSoup(r.text,'html.parser')
    allowlist = ['p','a']
    text_elements = [t for t in soup1.find('div', attrs={'data-area': 'body'}).find_all(text=True) if t.parent.name in allowlist]
    #temp1=[]
    #for text in soup.find_all('p'):
      #  temp=text.getText().strip('\n')
      #  temp = re.sub('&\w+;','',text_elements)
       # temp1.append(temp)
   # content.append(" ".join(temp))
    #text_elements= soup.find_all('p').getText().strip()
    text_elements=' '.join(text_elements)
    content.append(' '.join(text_elements.replace('\n', '').split()))
    img.append(soup1.find('picture').find('img')['src'])
    date.append(soup1.find('time', class_="timeformat")['datetime'])
    #author = soup.find(class_="author").span.get_text(strip=True)
    # text_elements = re.sub('&\w+;','',text_elements)

#print(news[0])
#print(link)
#print(title)
#print(content)
#print(img)
df= pd.DataFrame({'title':title,'link':link ,'content':content,'pubDate':date,'image_url':img})
#df.to_csv('world_news.csv')
```
